```python
import numpy as np
from scipy.stats import ttest_ind

# Extracting post-molt size and pre-molt size data
post_molt_size_data = df['post molt size']
pre_molt_size_data = df['pre molt size']

# Calculating Cohen's d
def calculate_cohens_d(group1, group2):
    mean_diff = np.mean(group1) - np.mean(group2)
    pooled_std = np.sqrt((np.std(group1, ddof=1)**2 + np.std(group2, ddof=1)**2) / 2)
    cohen_d = mean_diff / pooled_std
    return cohen_d

cohens_d = calculate_cohens_d(post_molt_size_data, pre_molt_size_data)

# Print the result
print(f"Cohen's d for Post Molt Size and Pre Molt Size: {cohens_d}")
```

```python
# Extracting pre-molt size and post-molt size data
pre_molt_size_data = df['pre molt size']
post_molt_size_data = df['post molt size']

# Calculating the observed test statistic (e.g., difference in means)
observed_statistic = np.mean(post_molt_size_data) - np.mean(pre_molt_size_data)

num_permutations = 1000

# Initializing an array to store permutation test statistics
perm_test_statistics = np.zeros(num_permutations)

# Permutation test
for i in range(num_permutations):
    # Concatenate the data and shuffle
    combined_data = np.concatenate((pre_molt_size_data, post_molt_size_data))
    np.random.shuffle(combined_data)

    # Calculate the test statistic for the permuted data
    perm_test_statistics[i] = np.mean(combined_data[:len(pre_molt_size_data)]) - np.mean(combined_data[len(pre_molt_size_data):])

# Calculating p-value
p_value = (np.abs(perm_test_statistics) >= np.abs(observed_statistic)).mean()

# Print the result
print(f"Observed Statistic: {observed_statistic}")
print(f"P-value: {p_value}")
```

```python
# Extracting pre-molt size and post-molt size data
pre_molt_size_data = df['pre molt size']
post_molt_size_data = df['post molt size']

# Performing t-test
t_statistic, p_value = ttest_ind(pre_molt_size_data, post_molt_size_data, equal_var=False)

# Displaying the results
print(f"T-Statistic: {t_statistic}")
print(f"P-Value: {p_value}")

# Check for statistical significance (common significance level is 0.05)
alpha = 0.05
if p_value < alpha:
    print("The mean difference is statistically significant.")
else:
    print("The mean difference is not statistically significant.")
```
